# [Importing](#Importing)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression 
from sklearn.metrics import accuracy_score,confusion_matrix,recall_score,precision_score
```

# [Data Loading](#data-loading)

```python
Diabetes=pd.read_csv("diabetes.csv")
Diabetes
```

# [Testing and Training](#Testing-and-Training)

```python
X=Diabetes.drop(columns=['Outcome'])
Y=Diabetes['Outcome']
```

```python
Y
```

```python
X
```

```python
X_train,X_test,Y_train,Y_test =train_test_split(X,Y,test_size=0.40,random_state=0)
```

```python
model = LogisticRegression(max_iter=1000)
model.fit(X_train,Y_train)
```

```python
Y_pred=model.predict(X_test)
```

```python
DF_test_predict= X_test
DF_test_predict["Y_test"]=Y_test
DF_test_predict["Y_pred"]=Y_pred
DF_test_predict
```

```python
CM=confusion_matrix(Y_test,Y_pred)
AS=accuracy_score(Y_test,Y_pred)
RS=recall_score(Y_test,Y_pred)
PS=precision_score(Y_test,Y_pred)
print(f"Accuracy Score= {AS}\nRecall Score= {RS}\nPrecision Score={PS}\nConfusion Matrix={CM}")
```

# [Data Visualization](#data-visualization)

```python
#Scatter Graph
sns.scatterplot(y='Glucose', x='Insulin', hue='Outcome', data=Diabetes, palette=['green','red'])
plt.title('Glucose vs BMI by Outcome')
plt.show()
```

```python
PregVSOut=Diabetes.groupby(["Pregnancies","Outcome"]).size().unstack(fill_value=0)
PregVSOut
```

```python
# Bar Graph
PregVSOut.plot(kind='bar')
plt.show()
```

```python
# Pie chart
plt.figure(figsize=(10,6))
labels = ['Diabetic', 'Non-Diabetic']
colors = ['#ff9999', '#66b3ff'] 
df=Diabetes['Outcome'].value_counts()
plt.pie(df, labels=labels, colors=colors, autopct='%.2f%%', startangle=140)
plt.title('Diabetic vs Non-Diabetic Population')
plt.show()
```

```python
#Histograph
Diabetes.hist(figsize=(10,9))
plt.suptitle("HISTOGRAPHS")
plt.show()
```

```python
plt.figure(figsize=(5,4))
sns.heatmap(CM,annot=True,fmt='d',cmap='Blues',xticklabels=["Non-Dibetic","Diabetic"],yticklabels=["Non-Dibetic","Diabetic"])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python

corr_matrix = Diabetes.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
