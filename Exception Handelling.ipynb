```python
a=10
b=0

try:
  print(a/b)
  L[8]
except ZeroDivisionError:
  print("B cannot be zero")
except NameError:
  print("Incorrect Identifier")
except IndexError:
  print("Trying to fetch the non-existing data")
finally:

  print("Code executed sucessfully")
```
