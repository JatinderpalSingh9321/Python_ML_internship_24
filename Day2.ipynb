```python
#prime number
def prime(a):
  if (a==1):
    print("its not a prime number")
  elif a>1:
      for i in range(2,a):
          if(a%i==0):
              print("not a prime number")
              break
          else:
              print("It is a prime number")
          return a

b=int(input("Enter a number :"))
prime(b)
```
