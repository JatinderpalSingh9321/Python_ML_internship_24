```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
import os
%matplotlib inline
```

```python
ti = pd.read_csv("/content/drive/MyDrive/titanic.csv")
ti
```

```python
#Graph of sex of people
mVSf=ti.groupby(["survived","sex"]).size().unstack(fill_value=0)
mVSf
```

```python
mVSf.plot(kind="bar",color=['pink','blue'])
plt.ylabel("Survived")
plt.xlabel("Sex")
plt.grid()
plt.show()
```

```python
#Graph of Alive or dead people
dt=ti.groupby(["who","alive"]).size().unstack(fill_value=0)
dt
```

```python
dt.plot(kind="bar", color=['r','g'])
plt.ylabel("Alive")
plt.xlabel("Sex")
plt.grid()
plt.show()
```

```python
#Graph of address of people
Address=ti.groupby(["sex","embark_town"]).size().unstack(fill_value=0)
Address
```

```python
Address.plot(kind="bar")
plt.ylabel("sex")
plt.xlabel("embark_town")
plt.grid()
plt.show()
```

```python
#Graph of people who are alone
family=ti.groupby(["alone","who"]).size().unstack(fill_value=0)
family
```

```python
family.plot(kind="bar")
plt.grid()
plt.show()
```
