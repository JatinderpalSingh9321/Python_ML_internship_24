create a code which sort car according to their price into 4 categories(0-10000 Budget, 10001-20000 Average, 20001-30000 Above Average, 30000 above Luxury)

```python

def car_class(C):
  if C == '?':
    return 'Invalid value'
  C=int(C)
  if 5000 <= C <= 10000:
    return 'Budget car'
  elif 10001 <= C <= 20000:
    return 'Medium car'
  elif 20001 <= C <= 30000:
    return 'Above Medium car'
  elif C > 30000:
    return 'Luxury car'
  else:
    return 'Invalid value'

auto['type'] = auto['price'].apply(car_class)
auto
```

```python
gg=auto.groupby(["make","type"]).size().unstack(fill_value=0)
gg
```

```python
gg.plot(kind="bar", stacked=True)
plt.show()
```
