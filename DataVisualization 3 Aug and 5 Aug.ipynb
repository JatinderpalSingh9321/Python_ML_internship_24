```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
sm=pd.read_csv('/content/drive/MyDrive/student_marks.csv')
```

```python
sm
```

```python
sm['Total']=sm[['English','Hindi','Maths','Science','SST']].sum(axis=1)
```

```python
sm
```

```python
sm['Percentage']=sm['Total']/5
```

```python
sm
```

**Grading**

A=100-80

B=80-60

C=60-50

D=50-40

E=40-30

F=0-30

```python
def clac_grade(P):
  if P<=100 and P>=80:
    return 'A'
  elif P<=80 and P>=60:
    return 'B'
  elif P<=60 and P>=50:
    return 'C'
  elif P<=50 and P>=40:
    return 'D'
  elif P<=40 and P>=30:
    return 'E'
  elif P<=30 and P>=0:
    return 'F'
  else:
    return 'X'

sm['Grade']=sm['Percentage'].apply(clac_grade)
sm
```
